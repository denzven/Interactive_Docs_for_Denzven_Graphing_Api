# Setting stuff up...

to use Denzven-Graphing-Api you will need a basic knowledge of python and also a bit about url encoding.. for this, we will use urllib.

## How to use it?

using my Api is really simple! it just needs a request to the api and it will return an image of the required graph.in this notebook i will be using python to do get make a request and get the output.

```python
import urllib.parse
```

importing urllib parse to parse our formula or to urllencode our input formula

```python
ApiBaseUrl = "https://denzven.pythonanywhere.com"
```

```python
ApiBaseUrl_FlatGraph = ApiBaseUrl + "/DenzGraphingApi/v1/flat_graph/test/plot"
```

defining the base url for the api... note: "v1/**flat_graph**/test/plot" this is the flat graph endpoint

```python
def ApiReqExample_FlatGraph(*input_params):
    #params = input_params.split(' ')
    params = input_params
    i = 0
    
    for e in params:
        if i == 0:
            e = urllib.parse.quote(e, safe='')
            ReqUrl_Flat = ApiBaseUrl_FlatGraph + f"?formula={e}"
            i += 1
        elif i != 0:
            ReqUrl_Flat = ReqUrl_Flat + f"&{e}"
    return(ReqUrl_Flat)
```

```python
url = ApiReqExample_FlatGraph('x+y','grid=1')
print(url)
```

this returns the required url for us. I'll display it using aiohttp and ipython display in this ipynb.

```python
import aiohttp
async with aiohttp.ClientSession() as s:
    async with s.get(url) as r:
        file = open("ApiReqExample_FlatGraph.png", "wb")
        file.write(await r.read())
        file.close()
```

```python
from IPython.display import Image
Image(filename = "ApiReqExample_FlatGraph.png")
```

a Beautiful graph indeed! But wait there's a ton more!
